```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import glob
from torchvision.utils import make_grid
from torchvision.io import read_image
import random
```

```python
path = '../../images/test_poses_contrast_npc40_background/'
imgs = sorted(glob.glob(path + '*.png'))
# imgs = imgs[:10]
imgs = random.sample(imgs, 10)
```

```python
len(imgs)
```

```python
imgs = [read_image(i) for i in imgs]
grid = make_grid(imgs, nrow=5, padding=10, pad_value=255.)
ndarr = grid.permute(1, 2, 0).cpu().numpy()
plt.imshow(ndarr)
```

```python
im = Image.fromarray(ndarr)
im.save(f'{path.split("/")[-2]}.png')
```
