```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import glob
from torchvision.utils import make_grid
from torchvision.io import read_image
import random
```

```python
path = '../../data/express80/'
imgs = sorted(glob.glob(path + '*.png'))
imgs = random.sample(imgs, 16)
```

```python
len(imgs)
```

```python
imgs = [read_image(i) for i in imgs]
grid = make_grid(imgs, nrow=8, padding=10, pad_value=255.)
ndarr = grid.permute(1, 2, 0).cpu().numpy()
plt.imshow(ndarr)
```

```python
im = Image.fromarray(ndarr)
im.save(f'{path.split("/")[-2]}.png')
```
