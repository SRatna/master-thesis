```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import glob
from torchvision.utils import make_grid
from torchvision.io import read_image
import random
```

```python
path = '../../data/BFM/test-dataset/*/test/img/*/'
imgs = sorted(glob.glob(path + '*.png'))
imgs = random.sample(imgs, 20)
```

```python
len(imgs)
```

```python
imgs = [read_image(i) for i in imgs]
grid = make_grid(imgs, nrow=10, padding=10, pad_value=255.)
ndarr = grid.permute(1, 2, 0).cpu().numpy()
plt.figure(figsize=(16, 16))
plt.imshow(ndarr)
```

```python
im = Image.fromarray(ndarr)
im.save(f'images/random-bfm-20.png')
```
