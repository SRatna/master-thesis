```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import glob
from torchvision.utils import make_grid
from torchvision.io import read_image
from torchvision import transforms
```

```python
def read_image(path):
    img = Image.open(path).convert('RGB').resize((224, 224), Image.ANTIALIAS)
    return transforms.ToTensor()(img)
```

```python
base_path = '../../new-notebooks/reco_images_80-model'
```

```python
best_ids = [28, 40, 2, 42, 0, 72, 76, 80] #94 96
```

```python
worst_ids = [4, 18, 30, 36, 46, 48, 58, 90] #92 84 56 54 
```

```python
good_ids = [6, 8, 14, 68, 24, 34, 96, 82] #0 68
```

```python
img = [f'{base_path}/outputs-masked/img{str(i).zfill(4)}.jpg' for i in best_ids]
rec_img = [f'{base_path}/reco{str(i).zfill(4)}.png' for i in best_ids]
imgs = img + rec_img
imgs = [read_image(i) for i in imgs]
grid = make_grid(imgs, nrow=8, padding=10, pad_value=1)
ndarr = grid.mul(255).clamp(0, 255).byte().permute(1, 2, 0).cpu().numpy()
plt.figure(figsize=(16, 16))
plt.imshow(ndarr)
```

```python
im = Image.fromarray(ndarr)
im.save(f'images/reco_best.png')
```
